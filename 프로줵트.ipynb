```python
def Tlqkf(data):
    data['상장주식수대비거래량'] = (data['거래량']/data['상장주식수'])*100
    data['시가총액대비거래대금'] = (data['거래대금']/data['시가총액'])*100
    idx = data[data['시가']==0].index
    data = data.drop(idx)
    data = data.sort_values('상장주식수대비거래량',ascending=True)
    data.reset_index(drop=True, inplace=True)
    data.set_index
    return data
```
